# Data Preparation

```python
#supprimer les features corrélés
cols_to_drop = ['histogram_mode', 'histogram_mean', 'histogram_min']
df.drop(cols_to_drop, axis=1, inplace=True)
```

```python
#afficher le nombre de colonnes après la suppression de features corrélés
df.shape
```

```python
#attribuer les valeurs aux features en tant que X et à la variable target en tant que y
X=df.drop(["fetal_health"],axis=1)
y=df["fetal_health"]
```

```python
#oversampler le dataSet avec la méthode smote
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=0)
X_smote, y_smote = smote.fit_resample(X, y)
balenced_data = pd.concat([X_smote,y_smote],axis=1)
X=X_smote
y=y_smote
```

```python
#afficher le diagramme circulaire après l'oversampling du DataSet
s0 = np.sum(balenced_data["fetal_health"] == 1)
s1 = np.sum(balenced_data["fetal_health"] == 2)
s2 = np.sum(balenced_data["fetal_health"] == 3)

fig1, ax1 = plt.subplots()
ax1.pie([s0, s1, s2],
        labels=['NORMAL', 'SUSPECT', 'PATHOLOGICAL'], 
        autopct='%1.0f%%',
        startangle=90,
        colors = [ '#483D8B','#4682B4','#87CEFA'])
ax1.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#La normalisation(Scalling)
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
#Configurer un scaler standard pour les features
col_names = list(X.columns)
s_scaler = preprocessing.StandardScaler()
X_df= s_scaler.fit_transform(X)
X_df = pd.DataFrame(X_df, columns=col_names)   
```

```python
#Enregistrer le modèle de scaling
import pickle
file_path = 'Classification_app/scaler_model.pkl'
pickle.dump(s_scaler, open(file_path, 'wb'))
```

```python
#Traçage de la boite à moustaches
boxplot = X_df.boxplot(figsize=(15,4),rot=90)
```

```python
#Séparation des données : Train(80%)/Test(20%)
from sklearn.model_selection import train_test_split
X1_train,X1_test,y1_train,y1_test=train_test_split(X_df,y,test_size=0.20,random_state=0)
```

```python
#Modélisation
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.ensemble import GradientBoostingClassifier

pipeline_svm = Pipeline([('svm_classifier',SVC(random_state=0))])

pipeline_lr = Pipeline([('lr_classifier',LogisticRegression(random_state=0))])

pipeline_dt = Pipeline([('dt_classifier',DecisionTreeClassifier(random_state=0))])

pipeline_rf = Pipeline([('rf_classifier',RandomForestClassifier(random_state=0))])

pipeline_knn = Pipeline([('knn_classifier',KNeighborsClassifier(n_neighbors=3))])

pipeline_gb = Pipeline([('gb_classifier',GradientBoostingClassifier(random_state=0))])

# Liste des modèles
pipelines = [pipeline_svm,pipeline_lr, pipeline_dt, pipeline_rf, pipeline_knn,pipeline_gb]

# Dictionnaire des pipelines et des types de classificateurs pour faciliter la référence
pipe_dict = {
             0: 'Support Vector Machine',
             1: 'Logistic Regression', 
             2: 'Decision Tree', 
             3:'RandomForest', 
             4: 'KNN',
             5:'Gradient Boosting'
            }
#Entrainement des modèles
for pipe in pipelines:
    pipe.fit(X1_train, y1_train)
```

```python
#la validation croisée avec la techniques KFold
from sklearn.metrics import f1_score,make_scorer
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold 
cross_val = KFold(n_splits=5)#assure que chaque pli contient une proportion similaire d'exemples de chaque classe
scorer = make_scorer(f1_score, average='micro')
for i, model in enumerate(pipelines):
    f1_scores = cross_val_score(model, X1_train, y1_train, scoring=scorer, cv=cross_val)
    print(pipe_dict[i],": f1-score : ",(f1_scores.mean()))
```

```python
#Test et évaluation des modèles en utilisant f1_score et MCC
from sklearn.metrics import f1_score, matthews_corrcoef
for i, model in enumerate(pipelines):
    ypred = model.predict(X1_test)
    model_score = f1_score(y1_test, ypred,average='weighted')
    MCC = matthews_corrcoef(y1_test, ypred)
    print(pipe_dict[i],": f1-score : ",model_score," MCC : ",MCC,"\n")
```

```python
#afficher le rapport de classification des mdèles
from sklearn.metrics import classification_report
for i, model in enumerate(pipelines):
    y_pred = model.predict(X1_test)
    print("----------------%s---------------- \n" % (pipe_dict[i]))
    print(classification_report(y1_test, y_pred))
```

```python
#afficher la matrice de confusion des mdèles
from scikitplot.metrics import plot_confusion_matrix
for i, model in enumerate(pipelines):
    y_pred = model.predict(X1_test)
    plot_confusion_matrix(y1_test,y_pred,figsize=(3,2),normalize=False)
    plt.title(pipe_dict[i],size=10)
    plt.show()
```

# Optimiser Random Forest, Decision Tree et Gradient Boosting


# 1-Random Forest 

```python
#afficher les paramètres et leurs valeurs actuelles pour Random Forest
pipeline_rf['rf_classifier'].get_params()
```

```python
#Optimisation du modèle :Random Forest avec RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
hyper_params_rf = { 
    'n_estimators': [100,150, 200,500,700,900],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,6,8,12,14,16],
    'criterion' :['gini', 'entropy'],
    'n_jobs':[-1,1,None]
}
grid_rf = RandomizedSearchCV(pipeline_rf['rf_classifier'], hyper_params_rf, scoring='f1_weighted', cv=5, n_iter=40)
grid_rf.fit(X1_train, y1_train)
print(grid_rf.best_params_)
```

```python
#Evaluation du modèle optimisé (Random Forest)
y_pred = grid_rf.predict(X1_test)
print('---------------Random forest---------------\n')
print('F1 : ', f1_score(y1_test, y_pred,average='weighted'))
print('MCC : ', matthews_corrcoef(y1_test, y_pred),'\n')
#afficher le rapport de classification
print(classification_report(y1_test, y_pred))
#afficher la matrice de confusion
plot_confusion_matrix(y1_test,y_pred,figsize=(3,2),normalize=False)
plt.title('Matrice confusion',size=10)
plt.show()
```

# 2-Decision Tree

```python
#afficher les paramètres et leurs valeurs actuelles pour Decision Tree
pipeline_dt['dt_classifier'].get_params()
```

```python
#Optimisation du modèle : Decision Tree avec RandomizedSearchCV

hyper_params_dt ={
              'max_features': ['auto', 'sqrt', 'log2'],
              'ccp_alpha': [0.1, .01, .001],
              'max_depth' : [5, 6, 7, 8, 9],
              'criterion' :['gini', 'entropy']
             }
grid_dt = RandomizedSearchCV(pipeline_dt['dt_classifier'], hyper_params_dt, scoring='f1_weighted', cv=5, n_iter=40)
grid_dt.fit(X1_train, y1_train)
print(grid_dt.best_params_)
```

```python
#Evaluation du modèle optimisé (Decision Tree)
y_pred = grid_dt.predict(X1_test)
print('---------------Decision Tree---------------\n')
print('F1 : ', f1_score(y1_test, y_pred,average='weighted'))
print('MCC : ', matthews_corrcoef(y1_test, y_pred),'\n')
#afficher le rapport de classification
print(classification_report(y1_test, y_pred))
#afficher le rapport de classification
plot_confusion_matrix(y1_test,y_pred,figsize=(3,2),normalize=False)
plt.title('Matrice confusion',size=10)
plt.show()
```
